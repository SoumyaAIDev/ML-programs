```python
import pandas as pd
import numpy as np
df =pd.read_excel("sales.xlsx")
df
```

```python
df["Cost"] = df["Weekly_Sales"] * 0.8   # assume 80% sales
```

```python
df["Percentage Change"] = ((df["Weekly_Sales"] - df["Cost"]) / df["Cost"]) * 100   #adding new cost to dataset
```

```python
print(df)
```

```python
df.to_excel("updated_sales.xlsx", index=False)

print("Updated data saved to 'updated_sales.xlsx'.")
```
